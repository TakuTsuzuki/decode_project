パディング　幅：pad_width

```python
def padwithtens(vector, pad_width, iaxis, kwargs):
    vector[:pad_width[0]] = 0
    vector[-pad_width[1]:] = 0
    return vector
```

蛍光画像からの，マスク生成

1, The erosion makes the object in white smaller.
2, dilatation makes the object in white bigger.


```python
def make_mask(image, closing_kernel=5, erosion_kernel=6, dilation_kernel=7):
    #calc thereshold automaticaly(http://scikit-image.org/docs/dev/api/skimage.filters.html)
    thresh = skimage.filters.threshold_otsu(image) 
    # morphology: closing → erosion →dilation
    mask = skimage.morphology.closing(image > thresh/2, square(closing_kernel)) #closing
    mask = skimage.morphology.erosion(mask, square(erosion_kernel)) #erosion
    mask = skimage.morphology.dilation(mask, square(dilation_kernel)) #dilation
    # border removal
    mask = clear_border(mask)
    return mask
```

各マスクに対して，番号を割り振る．

```python
def make_label(mask, gaussian_kernel=3):
    #caluculate distance on each pixcells 
    #apply gaussian blur to distance results
    distance = skimage.filters.gaussian(ndimage.distance_transform_edt(mask), gaussian_kernel)
    # detect maximum value on distance 
    local_maxi = peak_local_max(distance, indices=False, footprint=np.ones((3, 3)), labels=mask)
    markers = skimage.measure.label(local_maxi)
    # watershed
    labels_ws = skimage.morphology.watershed(-distance, markers, mask=mask)
    return labels_ws
```

crop and save images

```python
def crop_and_save(dic, savepath, size=90, cutoff=1800, pad=50):
    j=0
    area = np.array([])
    #image_dic = {}
    for impath in dic:
        image=io.imread(impath)
        mask = make_mask(image[0])
        labels = make_label(mask)
        
        image = np.pad(image, pad, padwithtens)[pad:pad+4]
        mask = np.pad(mask, pad, padwithtens)
        labels = np.pad(labels, pad, padwithtens)
        for i in xrange(len(regionprops(labels))):
            print j
            # skip small images
            if regionprops(labels)[i].area < cutoff:
                continue
            j=j+1
            
            # Record area size of each cell
            area = np.append(area, regionprops(labels)[i].area)
            # aplly mask to image 
            minr, minc, maxr, maxc = (regionprops(labels)[i]).bbox
            arr = (labels == i+1)*image
            # cropping!
            arr = arr[:,(maxr+minr-size)*0.5:(maxr+minr+size)*0.5,(maxc+minc-size)*0.5:(maxc+minc+size)*0.5]
            np.save(savepath+str(j)+".npy",  arr)
            #image_dic[ j ]=arr
            
        
    return area
```
